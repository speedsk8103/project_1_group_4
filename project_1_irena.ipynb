```python
print(happiness_2015_df.columns)
```

```python
# Analysis on 2015 data

gdp_2015_data = happiness_2015_df['Economy (GDP per Capita)'].values

gdp_mean_2015 = np.mean(gdp_2015_data)
gdp_std_2015 = np.std(gdp_2015_data)

print(f"The mean GDP per capita from 2015 was {gdp_mean_2015}")
print(f"The standard deviation of GDP per capita from 2015 was {gdp_std_2015}")
```

```python
gdp_2015_distribution = np.random.normal(gdp_mean_2015, gdp_std_2015, 1000)
plt.hist(gdp_2015_distribution, bins=30)
plt.xlabel('GDP per capita')
plt.ylabel('Frequency')
plt.title('Normal Distribution of GDP per Capita in 2015')
plt.savefig('gdp_2015_distribution.png')

plt.show()
```

```python
# Show correlation between happiness score and GDP per capita

x = happiness_2015_df['Happiness Score']
y = happiness_2015_df['Economy (GDP per Capita)']
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_title("Happiness Score and GDP per Capita 2015")
ax.set_xlabel("Happiness Score")
ax.set_ylabel("GDP per Capita")
slope, intercept, r_value, p_value, std_err = linregress(x, y)
line = slope * x + intercept
ax.plot(x, line, color='r')
r_squared = r_value ** 2
ax.text(0.05, 0.95, f"R-squared = {r_squared:.3f}", transform=ax.transAxes)
plt.savefig('gdp_2015_scatterline.png')
plt.show()
```

```python
# Correlation matrix

corr_matrix_2015 = happiness_2015_df.corr()
fig, ax = plt.subplots(figsize=(8, 8))
im = ax.matshow(corr_matrix_2015, cmap='coolwarm')
ax.set_xticks(range(len(corr_matrix_2015.columns)))
ax.set_yticks(range(len(corr_matrix_2015.columns)))
ax.set_xticklabels(corr_matrix_2015.columns)
ax.set_yticklabels(corr_matrix_2015.columns)
plt.xticks(rotation=45, ha='left')
fig.colorbar(im)
plt.show()
```
