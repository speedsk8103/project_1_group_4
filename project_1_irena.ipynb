```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import geopandas as gpd
```

```python
# Read csv files and import into dataframes
file_1 = "2015.csv"
happiness_2015_df = pd.read_csv(file_1)

# Clean up the dataframe to rename data and match to all other years

happiness_2015_df = happiness_2015_df.drop(columns=['Region'])
happiness_2015_df = happiness_2015_df.rename(columns={'Economy (GDP per Capita)': 'GDP per Capita'})
happiness_2015_df = happiness_2015_df.drop(columns=['Standard Error'])
happiness_2015_df = happiness_2015_df.drop(columns=['Dystopia Residual'])


happiness_2015_df.head()
```

```python
# Read csv files and import into dataframes
file_2 = "2016.csv"
happiness_2016_df = pd.read_csv(file_2)
# Clean up the dataframe to rename data and match to all other years

happiness_2016_df = happiness_2016_df.drop(columns=['Region'])
happiness_2016_df = happiness_2016_df.rename(columns={'Economy (GDP per Capita)': 'GDP per Capita'})
happiness_2016_df = happiness_2016_df.drop(columns=['Lower Confidence Interval','Upper Confidence Interval'])
happiness_2016_df = happiness_2016_df.drop(columns=['Dystopia Residual'])


happiness_2016_df.head()
```

```python
# Read csv files and import into dataframes
file_3 = "2017.csv"
happiness_2017_df = pd.read_csv(file_3)
# Clean up the dataframe to rename data and match to all other years

happiness_2017_df = happiness_2017_df.rename(columns={
    'Happiness.Rank': 'Happiness Rank',
    'Happiness.Score': 'Happiness Score',
    'Economy..GDP.per.Capita.': 'Economy (GDP per Capita)',
    'Health..Life.Expectancy.': 'Health (Life Expectancy)',
    'Trust..Government.Corruption.': 'Trust (Government Corruption)',
    'Dystopia.Residual': 'Dystopia Residual'
})
happiness_2017_df = happiness_2017_df.rename(columns={'Economy (GDP per Capita)': 'GDP per Capita'})
happiness_2017_df = happiness_2017_df.drop(columns=['Whisker.high','Whisker.low'])
happiness_2017_df = happiness_2017_df.drop(columns=['Dystopia Residual'])

happiness_2017_df.head()
```

```python
# Analysis on 2015 data

gdp_2015_data = happiness_2015_df['GDP per Capita'].values

# Find mean and standard deviation for the GDP per Capita
gdp_mean_2015 = np.mean(gdp_2015_data)
gdp_std_2015 = np.std(gdp_2015_data)

print(f"The mean GDP per capita from 2015 was {gdp_mean_2015}")
print(f"The standard deviation of GDP per capita from 2015 was {gdp_std_2015}")
```

```python
# Find and plot the normal distribution of GDP per Capita and Happiness Score

gdp_2015_distribution = np.random.normal(gdp_mean_2015, gdp_std_2015, 1000)
plt.hist(gdp_2015_distribution, bins=30)
plt.xlabel('GDP per capita')
plt.ylabel('Frequency')
plt.title('Normal Distribution of GDP per Capita in 2015')
plt.savefig('gdp_2015_distribution.png')

plt.show()
```

```python
# Find correlation between happiness score and GDP per capita, and create a scatter plot 

x = happiness_2015_df['Happiness Score']
y = happiness_2015_df['GDP per Capita']
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_title("Happiness Score and GDP per Capita 2015")
ax.set_xlabel("Happiness Score")
ax.set_ylabel("GDP per Capita")

# Plot the linear regression

slope, intercept, r_value, p_value, std_err = linregress(x, y)
line = slope * x + intercept
ax.plot(x, line, color='r')
r_squared = r_value ** 2
ax.text(0.05, 0.95, f"R-squared = {r_squared:.3f}", transform=ax.transAxes)
plt.savefig('gdp_2015_scatterline.png')
plt.show()
```

```python
# Pearson correlation between GDP and happiness
gdp_happ_2015_corr = happiness_2015_df['Happiness Score'].corr(happiness_2015_df['GDP per Capita'], method='pearson')
print("The Pearson correlation for 2015 data:", gdp_happ_2015_corr)
```

```python
# Plot a correlation matrix GDP and happiness

corr_matrix_2015 = happiness_2015_df.corr()
fig, ax = plt.subplots(figsize=(8, 8))
im = ax.matshow(corr_matrix_2015, cmap='coolwarm')
ax.set_xticks(range(len(corr_matrix_2015.columns)))
ax.set_yticks(range(len(corr_matrix_2015.columns)))
ax.set_xticklabels(corr_matrix_2015.columns)
ax.set_yticklabels(corr_matrix_2015.columns)
plt.xticks(rotation=45, ha='left')
fig.colorbar(im)
plt.savefig('2015_corr_matrix')
plt.show()
```
