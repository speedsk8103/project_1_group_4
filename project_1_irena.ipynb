```python
# Analysis on 2015 data

gdp_2015_data = happiness_2015_df['GDP per Capita'].values

# Find mean and standard deviation for the GDP per Capita
gdp_mean_2015 = np.mean(gdp_2015_data)
gdp_std_2015 = np.std(gdp_2015_data)

print(f"The mean GDP per capita from 2015 was {gdp_mean_2015}")
print(f"The standard deviation of GDP per capita from 2015 was {gdp_std_2015}")
```

```python
# Find and plot the normal distribution of GDP per Capita and Happiness Score

gdp_2015_distribution = np.random.normal(gdp_mean_2015, gdp_std_2015, 1000)
plt.hist(gdp_2015_distribution, bins=30)
plt.xlabel('GDP per capita')
plt.ylabel('Frequency')
plt.title('Normal Distribution of GDP per Capita in 2015')
plt.savefig('gdp_2015_distribution.png')

plt.show()
```

```python
# Find correlation between happiness score and GDP per capita, and create a scatter plot 

x = happiness_2015_df['Happiness Score']
y = happiness_2015_df['GDP per Capita']
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_title("Happiness Score and GDP per Capita 2015")
ax.set_xlabel("Happiness Score")
ax.set_ylabel("GDP per Capita")

# Plot the linear regression

slope, intercept, r_value, p_value, std_err = linregress(x, y)
line = slope * x + intercept
ax.plot(x, line, color='r')
r_squared = r_value ** 2
ax.text(0.05, 0.95, f"R-squared = {r_squared:.3f}", transform=ax.transAxes)
plt.savefig('gdp_2015_scatterline.png')
plt.show()
```

```python
# Pearson correlation between GDP and happiness
gdp_happ_2015_corr = happiness_2015_df['Happiness Score'].corr(happiness_2015_df['GDP per Capita'], method='pearson')
print("The Pearson correlation for 2015 data:", gdp_happ_2015_corr)
```

```python
# Plot a correlation matrix GDP and happiness

corr_matrix_2015 = happiness_2015_df.corr()
fig, ax = plt.subplots(figsize=(8, 8))
im = ax.matshow(corr_matrix_2015, cmap='coolwarm')
ax.set_xticks(range(len(corr_matrix_2015.columns)))
ax.set_yticks(range(len(corr_matrix_2015.columns)))
ax.set_xticklabels(corr_matrix_2015.columns)
ax.set_yticklabels(corr_matrix_2015.columns)
plt.xticks(rotation=45, ha='left')
fig.colorbar(im)
plt.savefig('2015_corr_matrix')
plt.show()
```

```python
# Calculate the difference in ranking and happiness score between 2015 and 2019
all_data_df['Ranking Diff'] = all_data_df.groupby('Country')['Happiness Rank'].transform(lambda x: x.max() - x.min())
all_data_df['Score Diff'] = all_data_df.groupby('Country')['Happiness Score'].transform(lambda x: x.max() - x.min())

# Combine the two data frames into one
diff_sorted = pd.concat([all_data_df.sort_values(by='Ranking Diff', ascending=False)[['Country', 'Ranking Diff']].drop_duplicates(),
                         all_data_df.sort_values(by='Score Diff', ascending=False)[['Country', 'Score Diff']].drop_duplicates()])

# Sort the combined data frame by the absolute value of the difference
diff_sorted = diff_sorted.sort_values(by=diff_sorted.columns[-1], ascending=False)

# Print the top 5 countries with the biggest changes in ranking and happiness score
print('Top 5 Countries with Biggest Change in Ranking and Happiness Score:')
print(diff_sorted.head(5))
```
