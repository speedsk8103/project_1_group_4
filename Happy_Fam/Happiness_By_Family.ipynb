```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import pandas as pd
from scipy.stats import linregress
```

```python
fifteenth = "../Resources/2015.csv"
sixteenth = "../Resources/2016.csv"
seventeenth = "../Resources/2017.csv"
eighteenth = "../Resources/2018.csv"
nineteenth = "../Resources/2019.csv"

fifteenth_data = pd.read_csv(fifteenth)
sixteenth_data = pd.read_csv(sixteenth)
seventeenth_data = pd.read_csv(seventeenth)
eighteenth_data = pd.read_csv(eighteenth)
nineteenth_data = pd.read_csv(nineteenth)

fif_df = pd.DataFrame(fifteenth_data)
six_df = pd.DataFrame(sixteenth_data)
sev_df = pd.DataFrame(seventeenth_data)
eight_df = pd.DataFrame(eighteenth_data)
nine_df = pd.DataFrame(nineteenth_data)
```

## Happiness Score and Family scatter plot relationship

In the years 2015-2017 the R-squared was roughly the same being .54 meaining that the relationship  is considred moderate. However in the year 2019 the R-squared jumped to .604 giving us a stronger claim that there is a relationshiop between the two. I will also add that after 2017 family was replaced with social support and could suggest that the study could have encompassed additional data as in my personal opinion I family brings me happiness but I would also say that friends hopefully being considred in "family" could ahve contriburted to a stronger relationship between the two.

I also want to preface that it was unknown how the scores were made so I assmumed that anything above 1.00 is considered high and anything as low as a zero score would be rated at the lowest possible score.

```python
x = fif_df['Happiness Score']
y = fif_df['Family']

fig, ax = plt.subplots()

ax.scatter(x, y)

ax.set_title("Happiness Score and Family Relationship 2015")
ax.set_xlabel("Happiness Score")
ax.set_ylabel("Family")

slope, intercept, r_value, p_value, std_err = linregress(x, y)
line = slope * x + intercept
ax.plot(x, line, color='r')

r_squared = r_value ** 2
ax.text(0.05, 0.95, f"R-squared = {r_squared:.3f}", transform=ax.transAxes)

plt.show()

fig.savefig('Happiness_2015.png', dpi=300, bbox_inches='tight')
```

## Comparing Happiness Score and Family to see if they are normally distributed 

While the normal distribution of family data appears to be slightly skewed to the left, it is important to note that this skewness is not significant enough to undermine the validity of our analysis. The use of the full dataset allows us to capture the overall trend and assess the relationship between happiness score and family, despite the slight deviation from a perfectly symmetrical distribution. By comparing the R-squared values across different years, we can see that there is a moderate to strong relationship between happiness score and family/social support, which further supports the validity of our analysis. 

Therefore, the skewness in the normal distribution does not detract from the overall conclusion that there is a relationship between happiness score and family/social support. Not to mention that if the swekeness is left leaning it can also indicate that family has a significant effect on Happiness score as a majority of scores are on the rihgt indicating that family is important.

```python
family_data_5 = fif_df['Family']

family_data_5 = fif_df['Family']
family_mean_5 = np.mean(family_data_5)
family_std_5 = np.std(family_data_5)
family_normal_5 = np.random.normal(family_mean_5, family_std_5, size=len(family_data_5))

fig, ax = plt.subplots()
ax.hist(family_data_5, bins=30, density=True, alpha=0.5, label="Family Score")
ax.plot(np.sort(family_normal_5), 1/(family_std_5 * np.sqrt(2 * np.pi)) *
        np.exp(-(np.sort(family_normal_5) - family_mean_5)**2 / (2 * family_std_5**2)), label="Normal Distribution")
ax.legend()
ax.set_title("Normal Distribution of Family Score 2015")
ax.set_xlabel("Family")
ax.set_ylabel("Density")

fig.savefig('Happiness_ND_2015.png', dpi=300, bbox_inches='tight')
plt.show()
```

It's worth noting that the top countries are mostly located in Europe, while the bottom countries are mostly located in Africa. Furthermore, the consistent pattern of high family scores among the top 5 countries and low family scores among the bottom 5 countries suggests that family is indeed an important factor when it comes to happiness.
