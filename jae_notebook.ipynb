```python
#calculate mean of happiness score and trust in government columns for each year's dataset. include linear regression.
#2015

x_values = fif_df['Trust (Government Corruption)']
y_values = fif_df['Happiness Score']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
fig, ax = plt.subplots()
r_squared = rvalue ** 2
ax.text(.1, 0.95, f"R-squared = {r_squared:.3f}", transform=ax.transAxes)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values,edgecolors="black", facecolors="steelblue")
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.title("Trust of Government and Happiness Scores 2015")
plt.xlabel("Trust of Government")
plt.ylabel("Happiness Score")
plt.savefig('images/trust_2015.png')
# Correlation between Trust of Government/Happiness Score in 2015.
corr_2015 = st.pearsonr(x_values,y_values)
corr_2015
```
